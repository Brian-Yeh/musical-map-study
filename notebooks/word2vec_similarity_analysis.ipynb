## Find Similarity Scores by Country

```python
cities_sim['country_code'] = cities_sim.location.str[-2:]
cities_sim.head()
```

```python
import pycountry

def country_code_to_name(df):
    return pycountry.countries.get(alpha_2=df['country_code']).name


cities_sim['country_name'] = cities_sim.progress_apply(country_code_to_name, axis=1)
# fix for USA
cities_sim.loc[cities_sim['country_code'] == 'US', 'country_name'] = 'United States of America'
cities_sim.head()
```

### Crime: Top 5 Similarity Scores by Country

```python
country_scores = cities_sim.groupby('country_name').mean()
top5_crime = country_scores.sort_values(by='crime_cos', ascending=False)
top5_crime[:5]['crime_cos']
```

### Economy: Top 5 Similarity Scores by Country

```python
top5_econ = country_scores.sort_values(by='economy_cos', ascending=False)
top5_econ[:5]['economy_cos']
```
