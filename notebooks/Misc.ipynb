```python
playlists = pd.read_csv('playlists.csv')
playlists
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="musical_map_study")
```

```python
location = geolocator.geocode(playlists.iloc[0]['name'])
print((location.latitude, location.longitude))
```

```python
playlists['long'] = ""
playlists['lat'] = ""
playlists.head()
```

```python
playlists_1 = playlists[:500].copy()
```

```python
from tqdm._tqdm_notebook import tqdm_notebook
tqdm_notebook.pandas()

pbar = tqdm_notebook(total=len(playlists))

# for i, row in playlists.iterrows():
#     location = geolocator.geocode(row['name'])
#     if location is not None:
#         playlists.iloc[i]['lat'] = location.latitude
#         playlists.iloc[i]['long'] = location.longitude
#     pbar.update(1)

from geopy.extra.rate_limiter import RateLimiter
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
playlists_1['location'] = playlists_1['name'].progress_apply(geocode)

playlists_1
```

```python
playlists_1['point'] = playlists_1['location'].apply(lambda loc: tuple(loc.point) if loc else None)
```

```python
import time

pbar = tqdm_notebook(total=len(playlists)-2500)

for i, row in playlists[2500:].iterrows():
    time.sleep(1)
    location = geolocator.geocode(row['name'])
    if location is not None:
        playlists.iloc[i]['lat'] = location.latitude
        playlists.iloc[i]['long'] = location.longitude
    pbar.update(1)
```
