```python
import pandas as pd
import ast
```

```python
song_info = pd.read_csv('../data/song_info.txt')
song_info = song_info[song_info['lang'] == 'en']
song_info = song_info[['song_id', 'title', 'artist', 'lyrics', 'audio_features']]
```

```python
sentiment_scores = pd.read_csv('../data/sentiment_scores.txt')
song_info = song_info.merge(right=sentiment_scores, how='inner', on='song_id')
song_info.head()
```

```python
song_info['musical_valence'] = song_info.audio_features.map(lambda x: ast.literal_eval(x)['valence'])
```

```python
sample = song_info.sample(5)
for i, row in sample.iterrows():
    stats = '[musical_valence: %s, vader_score: %s, textblob_score: %s]'%(row['musical_valence'], row['vader_score'], round(row['textblob_score'],4))
    print('\n\n=== Sample', i, '===', row['title'], 'by', row['artist'], stats)
    print(row['lyrics'])
```

#### Examples to look at
* === Sample 9041: Some Feeling by Mild Orange === musical_valence: 0.579 vader_score: 0.9724
* === Sample 7930: Fading Away by Adam Naas === musical_valence: 0.268 vader_score: 0.9973
* === Sample 2314: I Wanna Get Better by Bleachers === musical_valence: 0.398 vader_score: 0.9976

### Conclusion
It would be more useful to get a line averaged VADER score instead of passing the entire lyrics at once.
